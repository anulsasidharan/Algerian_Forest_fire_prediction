### Model Training

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# drop the unwanted columns
df.drop(['day','month', 'year'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
# Encoding
df['Classes']= np.where(df['Classes'].str.contains('not fire') , 0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## indipendent and dependent variables
# We are taking FWI -Fire weather index as the dependent (y) variable
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
## Feature Selection based on correlaltion
X_train.corr()
```

```python
# Check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr, annot=True)
```

- Here we are checking the multi correlations. If two columns are highly correlated, they act like similar data points. 
- For example, 
    - a highly correlated to b 
    - b highly correlated to c
    - that meand a and c are highly correlated. 
    - Therefore we can drop b as a and c are highly correlated. 

- In such cases we can set a threshold for the collinearinty, if the correlation is above those threshold, we can drop those data from the datasets. 


```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
    
```

```python
# Threshold: It will be set by the domain experts
corr_features = correlation(X_train, 0.85)
```

```python
corr_features
```

```python
# Drop the deatures when correlation is morethan 0.85
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape, X_test.shape
```

### Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scled
```

### Box plot to understand the Effect of Standard scaler

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')

plt.subplot(1,2,2)
sns.boxplot(data=X_train_scled)
plt.title('X_train Before Scaling')
```

## 1. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg = LinearRegression()
linreg.fit(X_train_scled, y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error: ", mae)
print("R2 Score", score)
plt.title("Multiple linear regression plot")
plt.scatter(y_test, y_pred)
```

## 2. Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error: ", mae)
print("R2 Score", score)
plt.title("Lasso regression plot")
plt.scatter(y_test, y_pred)
```

## 3. Cross validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scled, y_train)
y_pred = lassocv.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.title("Cross Validation Lasso plot")
plt.scatter(y_test, y_pred)
```

## 4. Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge = Ridge()
ridge.fit(X_train_scled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.title("ridge regression plot")
plt.scatter(y_test, y_pred)
```

## 5. RidgeCV model

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv=5)
ridgecv.fit(X_train_scled, y_train)
y_pred = ridgecv.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.title("RidgeCV regression plot")
plt.scatter(y_test, y_pred)
```

## 6. Elasticnet Regression Model

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic = ElasticNet()
elastic.fit(X_train_scled, y_train)
y_pred = elastic.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.title("Elasticnet regression plot")
plt.scatter(y_test, y_pred)
```

## 7. ElasticNetCV

```python
from sklearn.linear_model import ElasticNetCV
elasticcv = ElasticNetCV()
elasticcv.fit(X_train_scled, y_train)
y_pred = elasticcv.predict(X_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.title("ElasticnetCV regression plot")
plt.scatter(y_test, y_pred)
```
